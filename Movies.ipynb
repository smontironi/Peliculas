Este notebook es para aplicar NPL usando el dataset __[IMDb Top 100 Movies](https://www.kaggle.com/datasets/themrityunjaypathak/imdb-top-100-movies)__  de Kaggle


```python
# importar datos
import pandas as pd

# para limpieza de datos
import string
```

```python
#leer el dataset
data = pd.read_csv('movies.csv')

#vemos la forma del dataset
data.shape
```

```python
#chequeamos el Head del dataset
data.head()

#vemos que en index 3 la pelicula Star Wars: Episode V ... tiene dos puntos :
```

```python
#removemos puntuaciones de movie_name
#declaramos una funcion
def punctuation_removal(messy_str):
    clean_list = [char for char in messy_str if char not in string.punctuation]
    clean_str = ''.join(clean_list)
    return clean_str
data['movie_name']= data['movie_name'].apply(punctuation_removal)
```

```python
#chequeamos el Head del dataset sin puntuaciones
data.head()

# Star Wars: Episode V ahora figura sin dos puntos :
```

```python
#chequeamos los primeros 30 resultados
data.head(30)
```

```python
#Aparecen numeros entre los titulos
#por ejemplo en index 22: 12 Angry Men
```

```python
#removemos numeros de movie_name
import re
def drop_numbers(list_text):
    list_text_new = []
    for i in list_text:
        if not re.search('\d', i):
            list_text_new.append(i)
    return ''.join(list_text_new)

data['movie_name'] = data['movie_name'].apply(drop_numbers)
```

```python
#chequeamos los primeros 30 resultados
data.head(30)

#en index 22: 12 Angry Men ya no figura el numero 12
```

```python
#removemos acentos y carcteres especiales
#chequeamos los primeros 40 resultados
data.head(40)
# vemos que en index 34 Lon The Professional posee un acento
```

```python
#removemos acentos
#creamos una funciones que remueve los acentos

import unicodedata
def remove_accented_chars(text):
    new_text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    return new_text

# aplicamos la funcion
data['movie_name'] = data.apply(lambda x: remove_accented_chars(x['movie_name']), axis = 1)
```

```python
#chequeamos los primeros 40 resultados
data.head(40)
#vemos que en index 34 Lon The Professional ya no aparece
```

```python
# creamos una funcion que remueve caracteres eseciales
def remove_special_characters(text):
    pat = r'[^a-zA-z0-9]' 
    return re.sub(pat, ' ', text)
 
# aplicamos la funcion
data['movie_name'] = data.apply(lambda x: remove_special_characters(x['movie_name']), axis = 1)
```

```python
#visualzamos palabras mas frecuentes
# importamos librerias
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words = 'english')
words = cv.fit_transform(data.movie_name)
sum_words = words.sum(axis=0)


words_freq = [(word, sum_words[0, idx]) for word, idx in cv.vocabulary_.items()]
words_freq = sorted(words_freq, key = lambda x: x[1], reverse = True)
frequency = pd.DataFrame(words_freq, columns=['Palabra', 'freq'])

plt.style.use('fivethirtyeight')
color = plt.cm.ocean(np.linspace(0, 1, 20))
frequency.head(20).plot(x='Palabra', y='freq', kind='bar', figsize=(15, 6), color = color)
plt.title("Palabras ms frecuentes en ttulos de peliculas")
plt.show()
```

```python
# nube de palabras

# importamos las librerias

from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer

```

```python
cv = CountVectorizer(stop_words = 'english')
words = cv.fit_transform(data.movie_name)
sum_words = words.sum(axis=0)

words_freq = [(word, sum_words[0, idx]) for word, idx in cv.vocabulary_.items()]
words_freq = sorted(words_freq, key = lambda x: x[1], reverse = True)

wordcloud = WordCloud(background_color = 'lightcyan', width = 2000, height = 2000).generate_from_frequencies(dict(words_freq))

plt.style.use('fivethirtyeight')
plt.figure(figsize=(10, 10))
plt.axis('off')
plt.imshow(wordcloud)
plt.title("Palabras ms frecuentes en ttulos de peliculas ", fontsize = 20)
plt.show()
```
